##1. Import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

##2. Read dataset

```python
dataset_path = './data/titanic_modified_dataset.csv'
df = pd.read_csv(dataset_path,
                 index_col='PassengerId')
```

```python
df
```

##3. Train test split and Scale

```python
dataset_arr = df.to_numpy().astype(np.float64)
X, y = dataset_arr[:, :-1], dataset_arr[:, -1]
```

```python
X = np.c_[np.ones((X.shape[0], 1)), X]
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y,
                                                    test_size=0.2,
                                                    random_state=2,
                                                    shuffle=True)
X_train, X_test, y_train, y_test = train_test_split(X_train,
                                                  y_train,
                                                  test_size=0.125,
                                                  random_state=2,
                                                  shuffle=True)
```

```python
normalizer = StandardScaler()
X_train[:, 1:] = normalizer.fit_transform(X_train[:, 1:])
X_val[:, 1:] = normalizer.transform(X_val[:, 1:])
X_test[:, 1:] = normalizer.transform(X_test[:, 1:])
```

##4. Define function

```python
def sigmoid(z):
  return 1 / (1 + np.exp(-z))
```

```python
def predict(X, theta):
  z = X @ theta
  y_hat = sigmoid(z)
  return y_hat
```

```python
def compute_loss(y, y_hat):
  y_hat = np.clip(y_hat, 1e-7, 1 - 1e-7)
  return (
      -y *
      np.log(y_hat) - (1 - y) *
      np.log(1 - y_hat)
  ).mean()
```

```python
def compute_gradient(X, y, y_hat):
  return X.T @ (y_hat - y) / y.size
```

```python
def update_theta(theta, gradient, lr):
  return theta - lr * gradient
```

```python
def compute_accuracy(X, y, theta):
  y_hat = predict(X, theta).round()
  acc = (y_hat == y).mean()
  return acc
```

##5. Initialize parameter

```python
lr = 0.01
epochs = 100
batch_size = 16

np.random.seed(2)
theta = np.random.uniform(
    size=X_train.shape[1]
)
```

```python
theta.shape, X_train.shape
```

##6. Training

```python
train_accs = []
train_losses = []
val_accs = []
val_losses = []
for epoch in range(epochs):
  train_batch_losses = []
  train_batch_accs = []
  val_batch_losses = []
  val_batch_accs = []
  for batch in range(0, X_train.shape[0], batch_size):
    X_i = X_train[batch:batch + batch_size]
    y_i = y_train[batch:batch + batch_size]

    y_hat = predict(X_i, theta)
    loss = compute_loss(y_i, y_hat)
    gradient = compute_gradient(X_i, y_i, y_hat)
    theta = update_theta(theta, gradient, lr)
    acc = compute_accuracy(X_train, y_train, theta)

    train_batch_losses.append(loss)
    train_batch_accs.append(acc)

    y_val_hat = predict(X_val, theta)
    val_loss = compute_loss(y_val, y_val_hat)
    val_acc = compute_accuracy(X_val, y_val, theta)

    val_batch_losses.append(val_loss)
    val_batch_accs.append(val_acc)

  train_batch_loss = sum(train_batch_losses) / len(train_batch_losses)
  val_batch_loss = sum(val_batch_losses) / len(val_batch_losses)
  train_batch_acc = sum(train_batch_accs) / len(train_batch_accs)
  val_batch_acc = sum(val_batch_accs) / len(val_batch_accs)

  train_accs.append(train_batch_acc)
  train_losses.append(train_batch_loss)
  val_accs.append(val_batch_acc)
  val_losses.append(val_batch_loss)

  print(f"\nEPOCH {epoch + 1}:\tTraining loss: {train_batch_loss:.3f}\tValidation loss: {val_batch_loss:.3f}")
```

##7. Visualize results

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 10))

ax[0, 0].plot(train_losses)
ax[0, 0].set(xlabel='Epoch', ylabel='Loss')
ax[0, 0].set_title('Training Loss')

ax[0, 1].plot(val_losses, 'orange')
ax[0, 1].set(xlabel='Epoch', ylabel='Loss')
ax[0, 1].set_title('Validation Loss')

ax[1, 0].plot(train_accs)
ax[1, 0].set(xlabel='Epoch', ylabel='Loss')
ax[1, 0].set_title('Train Accuracy')

ax[1, 1].plot(val_accs, 'orange')
ax[1, 1].set(xlabel='Epoch', ylabel='Loss')
ax[1, 1].set_title('Validation Accuracy')

plt.show()
```

##8. Evaluate

```python
val_set_accs = compute_accuracy(X_val, y_val, theta)
test_set_accs = compute_accuracy(X_test, y_test, theta)
print('Evaluation on validation and test set:')
print(f'Accuracy on validation set: {val_set_accs}')
print(f'Accuracy on test set: {test_set_accs}')
```
